### Train Test Split

#### Because there is imbalace in the DataSet (Churn column: more '0' values than '1' values) I will apply oversampling to fix this

```python
#x = df_1[['ContractRenewal', 'CustServCalls', 'DayMins', 'MonthlyCharge', 'OverageFee', 'RoamMins']]
x = df_1.iloc[:,:-1]
y = df_1.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.01,
                                                    random_state = 100)
```

```python
print('Before OverSampling, the shape of train_X: {}'.format(x_train.shape)) 
print('Before OverSampling, the shape of train_y: {} \n'.format(y_train.shape)) 

print("Before OverSampling, counts of label '1' in y_train is : {}".format(sum(y_train == 1))) 
print("Before OverSampling, counts of label '0' in y_train is : {} \n".format(sum(y_train == 0))) 

sm = SMOTE(random_state = 100) 
x_train_res, y_train_res = sm.fit_sample(x_train, y_train.ravel()) 

print('After OverSampling, the shape of train_X: {}'.format(x_train_res.shape)) 
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape)) 

print("After OverSampling, counts of label '1 in y_train is ': {}".format(sum(y_train_res == 1))) 
print("After OverSampling, counts of label '0'in y_train is: {}".format(sum(y_train_res == 0))) 
```

```python
new_x_train, new_x_test, new_y_train, new_y_test = train_test_split(x_train_res, y_train_res.ravel(), test_size = 0.3,
                                                    random_state = 100)
```

```python
print('Total number of 1 in new_y_train is: {}'.format(sum(new_y_train == 1)))
print('Total number of 0 in new_y_train is: {}'.format(sum(new_y_train == 0)))
```

```python
print('Total number of 1 in new_y_test is: {}'.format(sum(new_y_test == 1)))
print('Total number of 0 in new_y_test is: {}'.format(sum(new_y_test == 0)))
```

### Returning back to train test split:

```python
minmax= MinMaxScaler()

new_x_train = minmax.fit_transform(new_x_train)
new_x_test = minmax.transform(new_x_test)

```

## Using PyCaret to see which models are best to use

```python
pc_data.Churn.value_counts()
```

```python
models = setup(data = pc_data, target = 'Churn',normalize = True, normalize_method = 'minmax',transformation = True,
              verbose = False, html = True, train_size = 0.70,silent = True, session_id = 100, numeric_features = ['ContractRenewal', 'CustServCalls'])
```

```python
best9 = compare_models(n_select = 9)
```

```python
[print('\n','best9[{}] is: {}'.format(i,j),'\n'*3) for i,j in enumerate(best9)];
```

#### The PyCaret algorithm has selected the best 9 ML models to use

## Applying the models chosen by the PyCaret Algorithm

```python
def AccuracyReportTrain(y_true,y_pred):
    print('\n\nTrain Accuracy is: ',
          round(100*accuracy_score(y_true,y_pred),3),'%\n');
```

```python
def AccuracyReportTest(y_true,y_pred):
    print('\n\nTest Accuracy is: ',
          round(100*accuracy_score(y_true,y_pred),3),'%\n');
```

```python
def generateClassificationReportTrain(y_true,y_pred):
    print('\n\n','    '*6,'Train Classification  \n\n')
    print(class_report(y_true,y_pred));
    #print(confusion_matrix(y_true,y_pred));
    print('\n\nTrain Accuracy is: ',
          round(100*accuracy_score(y_true,y_pred),3),'%\n');
```

```python
def generateClassificationReportTest(y_true,y_pred):
    print('\n\n','    '*6,'Test Classification  \n\n')
    print(class_report(y_true,y_pred));
    #print(confusion_matrix(y_true,y_pred));
    print('\n\nTest Accuracy is: ',
          round(100*accuracy_score(y_true,y_pred),3),'%\n');
```

###  best9[0]: Extra Trees Classifier

#### Train

```python
etc = best9[0].fit(new_x_train,new_y_train)
ytrain_pred = etc.predict(new_x_train)

AccuracyReportTrain(new_y_train,ytrain_pred)

classes = [ "Unchurned","Churned"]
print('\n\n','    '*6,'Train Classification  \n\n')
disp = classification_report(etc, X_train = new_x_train, y_train = new_y_train, classes = classes,support=True)

classLabels = ['Not_Churned', 'Churned']
disp = plot_confusion_matrix(estimator=etc, X=new_x_train, y_true=new_y_train, cmap='Blues', values_format='20.0f', display_labels=classLabels)
disp.figure_.set_size_inches(w=8, h=8)
disp.ax_.set_title('Train Confusion Matrix', fontsize=13)
plt.grid()
```

#### Test

```python
ytest_pred = etc.predict(new_x_test)

AccuracyReportTest(new_y_test,ytest_pred)

classes = [ "Unchurned","Churned"]
print('\n\n','    '*6,'Test Classification  \n\n')
disp = classification_report(etc, X_train = new_x_train, y_train = new_y_train, X_test = new_x_test, y_test = new_y_test, classes = classes,support=True)

disp = plot_confusion_matrix(estimator=etc, X=new_x_test, y_true=new_y_test, cmap='Blues', values_format='20.0f', display_labels=classLabels)
disp.figure_.set_size_inches(w=8, h=8)
disp.ax_.set_title('Test Confusion Matrix', fontsize=13)
plt.grid()
plt.show()
```
